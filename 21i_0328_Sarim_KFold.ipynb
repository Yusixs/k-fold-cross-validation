```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris['data'])
df['target'] = iris['target']
df.columns = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Target"]
df
```

```python
def train_test_split(df, number_of_splits):

  start = 0
  k_splits = [pd.DataFrame(columns = df.columns) for i in range(number_of_splits)]

  for i in df['Target'].unique():

    singleLabelDf = df[df['Target'] == i]
    index = int(len(singleLabelDf) / number_of_splits)

    for j in range(number_of_splits):
        k_splits[j] = pd.concat([k_splits[j],singleLabelDf[start:start+index]], ignore_index = True)

    start += index

  return k_splits
```

```python
number_of_splits = 10

split = train_test_split(df, number_of_splits)
display(split)
```

```python
for index in range(number_of_splits):
  split = train_test_split(df, number_of_splits)
  test = split.pop(index)
  train = split

  train = pd.DataFrame(columns = df.columns)
  for i in split[:-1]:
    train = pd.concat([train, i])

  x_train, y_train = train.iloc[:,:-1], train.iloc[:,-1]
  x_test, y_test = test.iloc[:,:-1], test.iloc[:,-1]

  knn = KNeighborsClassifier(n_neighbors=3)
  knn.fit(x_train, y_train.astype('int'))

  y_pred = knn.predict(x_test)
  print(f"Fold {index}: {accuracy_score(y_test.astype('int'), y_pred)}")
```

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,:-1]
y = df.iloc[:,-1].astype("int")
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
print(f"Sklearn Train Test Split: {accuracy_score(y_test, y_pred)}")
```
